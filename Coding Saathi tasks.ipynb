```python
#task 2

import random
upper_range = 100
chances_allowed = 10
answer = random.randint(1, upper_range)
print("Welcome to the number guessing game!\n")
while True:
    level = input("Select difficulty level (easy, medium, hard): ").lower()
    if level in ["easy", "medium", "hard"]:
        break
    else:
        print("Invalid input. Please select either 'easy', 'medium', or 'hard'.")
        print()
if level == "easy":
    upper_range = 50
    chances_allowed = 20
elif level == "medium":
    upper_range = 100
    chances_allowed = 15
else:
    upper_range = 150
    chances_allowed = 10
answer = random.randint(1, upper_range)
print("\nI have chosen a number between 1 and", upper_range)
print("You have", chances_allowed, "guesses to find the correct number.")
for i in range(chances_allowed):
    userInput = input("Guess a number between 1 and " + str(upper_range) + ": ")
    guess = int(userInput)

    if guess == answer:
        print("Congratulations! You guessed the correct number. You win!")
        break

    elif guess < answer:
        print("The number is higher.")
    else:
        print("The number is lower.")

    if i == chances_allowed - 1:
        print("Sorry, you have run out of chances. You lose!")
        print("The correct number was:", answer)

    if abs(guess - answer) <= 10:
        print("You are warm!")
    elif abs(guess - answer) <= 20:
        print("You are getting warmer.")
    elif abs(guess - answer) <= 30:
        print("You're cold.")
    else:
         print("You are freezing.")
```

```python
import time

class User:
    def __init__(s, name, mail, password):
        s.name = name
        s.mail = mail
        s.password = password
        s.score = 0

    def log(s, password):
        if password == s.password:
            return True
        else:
            return False

    def profile(s, name, mail, password):
        s.name = name
        s.mail = mail
        s.password = password

    def select_answer(s, question):
        answer = input(question.question + " ")
        if answer == question.answer:
            s.score += question.score
            return True
        else:
            return False

    def timer(s, duration):
        for i in range(duration, 0, -1):
            print(i)
            time.sleep(1)
        s.submit_answers()

    def answers(s):
        print("Answers submitted. Your score is:", s.score)

    def logout(s):
        print("Logged out successfully.")

class Question:
    def __init__(s, question, answer, score):
        s.question = question
        s.answer = answer
        s.score = score

# create a user
user = User("John Doe", "johndoe@example.com", "password123")

# login
if user.log(input("Enter password: ")):
    print("Login successful.")
else:
    print("Invalid credentials.")

# update profile
user.profile("Jane Doe", "janedoe@example.com", "password456")
print("Profile updated.")

# create questions
question1 = Question("What is the best breakfast recognoised by WHO?", "idly", 40)
question2 = Question("Which country detected corona virus first?", "china", 25)
question3 = Question("What is the national animal of India?", "Elephant", 35)

# select answers
if user.select_answer(question1):
    print(" Nice Correct.")
else:
    print("Incorrect.")

if user.select_answer(question2):
    print(" Nice Correct.")
else:
    print("Incorrect.")

if user.select_answer(question3):
    print(" Nice Correct.")
else:
    print("Incorrect.")

user.answers()
user.logout()
```
